### binary sequences

```python
np.random.seed(0)
```

```python
def create_binary_sequence(n, p_first=0.5, p_later=[0.5, 0.5]):
    """
    create sequence of 1's and 0's of length n
    odds of first entry being 1 is p_first
    p_later = [a,b]. a = prob of i=1 given i-1=0, b = prob of i=1 given i=1=1.
    """
    sequence = []
    
    if np.random.random() > p_first:
        sequence.append(1)
    else:
        sequence.append(0)
    
    for _ in range(n-1):
        if np.random.random() < p_later[sequence[-1]]:
            sequence.append(1)
        else:
            sequence.append(0)
    
    return sequence
```

```python
ps = [
    [0.5, 0.5, 0.5],
    [0.5, 0.25, 0.75], # 75% odds of being same as previous one
    [0.5, 0.75, 0.25], # 75% odds of switching
    [0.5, 1,0], # guaranteed to switch everytimg
    [0.75, 0.75, 0.75],
    [0.25, 0.25, 0.25]
]
```

```python
for _ in range(50):
    p = ps[np.random.randint(0,5)]
    print(p)
    print(create_binary_sequence(n=30, p_first=p[0], p_later=[p[1], p[2]]))
```

### 2d randomness.

```python
np.random.seed(0)
```

```python
def create_2d_values(n, strategy):
    if strategy == 'uniform':
        return np.random.random(size=(n,)), np.random.random(size=(n,))
    
    if strategy == 'grid':
        n = round(np.sqrt(n))
        x = np.linspace(start=0, stop=1, num=n)
        y = np.linspace(start=0, stop=1, num=n)
        
        x = np.repeat(x,n)
        y = np.tile(y,n)
        
        return x,y
    
    if strategy == 'gridnoise':
        n = round(np.sqrt(n))
        
        x = np.linspace(start=0, stop=1, num=n)
        y = np.linspace(start=0, stop=1, num=n)
        
        x = np.repeat(x,n) + np.random.uniform(low=-1/n/3, high=1/n/3, size = n**2)
        y = np.tile(y,n) + np.random.uniform(low=-1/n/3, high=1/n/3, size = n**2)
        
        return x,y
    
    if strategy == 'normal':
        xy = np.random.randn(n,2)/8+0.5
        indices = np.logical_or(xy < 0, xy>1).sum(axis=1).astype(bool)
        xy = xy[~indices]
        return xy[:,0], xy[:,1]
```
